```python
#  Demonstration of Activation Functions: Sigmoid, ReLU, Tanh, and Softmax

import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 500)

# Define activation functions
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

# Plot: Sigmoid, ReLU, Tanh
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(x, sigmoid, color='blue', linewidth=2)
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)

plt.subplot(4, 1, 2)
plt.plot(x, relu, color='green', linewidth=2)
plt.title('ReLU (Rectified Linear Unit) Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(x, tanh, color='red', linewidth=2)
plt.title('Tanh (Hyperbolic Tangent) Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)

# Softmax Demonstration
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # stability trick
    return exp_z / exp_z.sum(axis=0)

# Simulate logits (raw outputs of 10 neurons)
logits = np.array([2.0, 1.0, 0.1, 0.5, 3.0, 0.2, 1.5, 0.8, 0.3, 2.5])
softmax_output = softmax(logits)

plt.subplot(4, 1, 4)
plt.bar(range(10), softmax_output, color='purple')
plt.title('Softmax Activation (10 Output Neurons)')
plt.xlabel('Neuron Index')
plt.ylabel('Probability')
plt.grid(True)

plt.tight_layout()
plt.show()

# --------------------  ANALYSIS --------------------
print("Activation Function Analysis:")
print("""
1 Sigmoid:
   - Range: (0, 1)
   - Used in binary classification (output layer)
   - Problem: vanishing gradient for large |x|

2 ReLU:
   - Range: [0, )
   - Fast convergence, avoids vanishing gradient
   - Problem: Dead neurons for negative inputs

3 Tanh:
   - Range: (-1, 1)
   - Zero-centered, better for hidden layers than sigmoid
   - Problem: Still vanishing gradients for large |x|

4 Softmax:
   - Converts logits (raw outputs) into probabilities that sum to 1
   - Used in output layer for multi-class classification
   - Example: 10 neurons  probability distribution across 10 classes
""")

# Show Softmax output clearly
for i, prob in enumerate(softmax_output):
    print(f"Neuron {i}: {prob:.3f}")
print(f"\nSum of probabilities = {softmax_output.sum():.2f}")
```
